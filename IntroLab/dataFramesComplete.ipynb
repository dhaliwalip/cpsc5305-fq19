```python
# Rename the column zip_code to be postal_code
users.rename(columns = {'zip_code': 'postal_code'}, inplace=True)
```

```python
users.columns
```

```python
# Create a new column salary; populate it with random numbers 
#   normally distributed with a mean of 100000 and a standard deviation of 1000
import numpy as np
np.random.normal(loc=100000, scale=1000, size = users.shape[0])
users['salary'] = np.random.normal(loc=100000, scale=1000, size = users.shape[0])
```

```python
users.columns
```

```python
users.salary = users.salary.map(lambda x: int(x))
```

```python
users.salary
```

```python
# Create the column dollars_per_year which is salary divided by age
users['dollars_per_year'] = users.salary / users.age
```

```python
# Drop the column dollars_per_year
users.drop(['dollars_per_year'], axis=1, inplace=True)
```

```python
# Compute average salary by occupation
users.groupby('occupation').salary.mean()
```

```python
# Change the coding of gender to 0 for male, 1 for female
users['is_male'] = users.gender.map({'F':0, 'M':1})
```

```python
users[['gender', 'is_male']]
```

```python
# Change the data type of zip_code to integer
users.postal_code.astype('float')
```

```python
users.postal_code.min()
```

```python
users.postal_code = users.postal_code.convert_objects(convert_numeric=True)
```

```python
users.postal_code.max()
```

```python
# Create a new data frame with a random sample of 50% of the rows in the original
mask = np.random.rand(users.shape[0]) <= 0.5
sampled_users = users[mask]
sampled_users.shape
```

```python
# Write the new data frame to a file user_sample.csv --
#  make it comma delimited, and no header line
sampled_users.to_csv('sampled_users.csv', header=False, sep=',')
```
